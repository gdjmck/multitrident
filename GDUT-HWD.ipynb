```python
import os
import glob
import cv2
```

```python
sizes = {}
img_folder = '/home/chk/GDUT-HWD/JPEGImages/'
for f in os.listdir(img_folder):
    img = cv2.imread(img_folder + f)
    size = img.shape[:2]
    if size not in sizes.keys():
        sizes[size] = 1
    else:
        sizes[size] += 1
```

```python
import xml.etree.ElementTree as ET
```

```python
annos = '/home/chk/GDUT-HWD/Annotations/'
labels = []
for f in os.listdir(annos):
    target = ET.parse(os.path.join(annos, f))
    for obj in target.iter('object'):
        label = obj.find('name').text
        if label not in labels:
            labels.append(label)
```
